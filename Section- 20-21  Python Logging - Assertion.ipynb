```python
import logging
logging.basicConfig(filename='mylog13',level=20, format='%(asctime)s:%(levelname)s:%(message)s',datefmt='%d/%m/%Y %I:%M:%S %p')

logging.info('A new request came')
try:
  x=int(input('Enter first number:'))
  y=int(input('Enter second number:'))
  print('The Result:',x/y)
except ZeroDivisionError as msg:
  print('Cannot divide with zero')
  logging.exception(msg)
except ValueError as msg:
  print('Provide int value only')
  logging.exception(msg)

logging.info('Request processing complete')
```

```python
#376 DEMO PROGRAM TO PERFORM DEBUGGING BY USING ASSERTION
def squareIt(x):
  return x*x

assert squareIt(2)==4,'The square of 2 should be 4'    #It will show AssertionError when the provided condition is false
assert squareIt(3)==9,'The square of 3 should be 9'
assert squareIt(4)==16,'The aquare of 4 should be 16'
print(squareIt(2))
print(squareIt(3))
print(squareIt(4))
```
