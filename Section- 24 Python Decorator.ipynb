```python
#412 Decorator demo
def decor_for_add(func):
  def inner(a,b):
    print('#'*30)
    print('The Sum:',end='')
    func(a,b)
    print('#'*30)
  return inner

@decor_for_add
def add(a,b):
  print(a+b)

add(10,20)
```

```python
#413 A
def decor(func):
  def inner(name):
    names=['CM','PM','Minister']
    if name in names:
      print('#'*50)
      print('Hello {} you are very important for us..'.format(name))
      print('Very Very Good Morning!!!')
      print('#'*50)
    else:
      func(name)
  return inner

@decor
def wish(name):
  print('Good Morning',name)

wish('Durga')
wish('Sunny')
wish('CM')

```

```python
def smart_division(func):
  def inner(a,b):
    if b==0:
      print('Hello Stupid ,How we can divide with ZERO!!!')
    else:
      func(a,b)
  return inner
@smart_division
def division(a,b):
  print(a/b)

division(10,2)
division(10,0)
  
```

```python
def decor(func):
  def inner(name):
    if name.lower()=='sunny':
      print('#'*50)
      print('Hello {} you are very important for us..'.format(name))
      print('Very Very Good Morning!!!')
      print('#'*50)
    else:
      func(name)
  return inner

def wish(name):
  print('Good Morning',name)
decorated_wish=decor(wish)

wish('sunny')
decorated_wish('Sunny')
```

```python
#417 Demo example program-2
def decor1(func):
  def inner1():
    x=func()    #num() will be called and consider return value
    return x*x
  return inner1
def decor2(func):
  def inner2():
    x=func()
    return 2*x
  return inner2
  
@decor2
@decor1
def num():
  return 20

print(num())
```
