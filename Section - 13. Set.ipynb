```python
#184 union() intersection() defference() symmetric_difference()
#UNION
s1={10,20,30,40}
s2={30,40,50,60}
s3=s1.union(s2)
print(s3)
#or
s3=s1|s2
print(s3)
#INTERSECTION
s4=s1.intersection(s2) 
print(s4)
#or
s4=s1&s2
print(s4)
#DIFFERENCE
s5=s1.difference(s2)
print(s5)
#or
s5=s1-s2
print(s5)
#SYMMETRIC_DIFFERENCE
s6=s1.symmetric_difference(s2)
print(s6)
#or
s6=s1^s2
print(s6)
```
